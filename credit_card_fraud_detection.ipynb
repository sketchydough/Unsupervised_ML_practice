```python
import pandas as pd
from sklearn.ensemble import IsolationForest
from sklearn.preprocessing import StandardScaler
```

```python
df= pd.read_csv("creditcard.csv")  
```

```python
df.head()
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
df.describe()
```

```python
X = df.drop(columns=['Time', 'Class'])
```

```python
#scaling features
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

```python
#fitting into isolation forest
iso_forest = IsolationForest(n_estimators=100, contamination=0.01, random_state=42)
iso_forest.fit(X_scaled)
```

```python
#predicting anomalies
#-1 for anomaly (i.e. fraud), 1 for normal
pred_iso = iso_forest.predict(X_scaled)
```

```python
#adding predictions to dataframe to visualize
df['anomaly'] = pred_iso
df['anomaly'] = df['anomaly'].map({1: 0, -1: 1})  # 1 = fraud, 0 = normal
```

```python
df['anomaly'].value_counts()
```

```python
df.head()
```

```python
from sklearn.metrics import classification_report, confusion_matrix
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
y = df['Class']  # Only for evaluation
confusion_matrix(y, pred_iso)
```

```python
#graphs
sns.countplot(x='anomaly', data=df)
plt.title('Number of Normal vs Fraud Transactions')
plt.show()
```

```python
#PCA for 2d visualization
from sklearn.decomposition import PCA

pca = PCA(n_components=2)
X_pca = pca.fit_transform(X_scaled)
plt.figure(figsize=(10,6))
plt.scatter(X_pca[:,0], X_pca[:,1], c=pred_iso, cmap='coolwarm', alpha=0.6)
plt.title('Transaction Clusters (Red = Fraud, Blue = Normal)')
plt.xlabel('PCA 1')
plt.ylabel('PCA 2')
plt.show()
```
